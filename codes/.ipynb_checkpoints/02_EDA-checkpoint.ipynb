```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression, LogisticRegression
```

```python
rats = pd.read_csv('../datasets/rats.csv')
```

```python
rat_sightings = pd.read_csv('../datasets/rat_sightings.csv')
```

```python
bbl_df = pd.read_csv('../datasets/bbl_df.csv')
```

```python
rats['inspection_date'] = pd.to_datetime(rats['inspection_date'])
rats['approved_date'] = pd.to_datetime(rats['approved_date'])
```

```python
# Let's scoop lat/long up in a matrix so we can use them easily
geo = rats[['longitude', 'latitude']]
geo.head()
```

```python
# Let's cluster our observations by latitude/longitude
km = KMeans(n_clusters=20)
km.fit(geo)
```

```python
plt.figure(figsize=(8, 12))
plt.scatter(rats.longitude, rats.latitude, c=km.labels_, s=1, cmap="tab20",alpha=.05);
```

```python
rats.sort_values(by = 'inspection_date', ascending = True, inplace = True)
```

```python
test = rats['inspection_date'][0]
```

```python
rats['year'] = pd.DatetimeIndex(rats['inspection_date']).year
```

```python
rats['month'] = pd.DatetimeIndex(rats['inspection_date']).month
```

```python
rats.head()
```

```python
rats['year_month'] = pd.to_datetime(rats['inspection_date']).dt.to_period('M')
```

```python
rats.head()
```

```python
rats = pd.concat([rats, pd.get_dummies(rats['inspection_type'])], axis = 1)
```

```python
rats.head()
```

```python
df = pd.merge(rats, bbl_df, left_on=rats['bbl'], right_on=bbl_df['bbl'], how='left')
df = df.dropna()
df['bbl_y'] = df['bbl_y'].astype(int)
df['zipcode'] = df['zipcode'].astype(int)
df.head()
```

```python
df.shape
```

```python
bbl_df.head()
```

```python
rat_sightings['created_date'] = pd.to_datetime(rat_sightings['created_date'])
rat_sightings['year'] = pd.DatetimeIndex(rat_sightings['created_date']).year
rat_sightings['month'] = pd.DatetimeIndex(rat_sightings['created_date']).month
rat_sightings['year_month'] = pd.to_datetime(rat_sightings['created_date']).dt.to_period('M')
rat_sightings.head()
```

```python
rat_sightings['sighting'] = 1
sightings = rat_sightings.groupby(by = ['incident_zip', 'year_month'])[['sighting']].sum().reset_index()
sightings.head()
```

```python
sightings['zip_year_month'] = sightings['incident_zip'].astype(str)+' '+sightings['year_month'].astype(str)
sightings.head()
```

```python
bbl_year_month = df.groupby(by = ['zipcode', 'year_month'])[['BAIT', 'CLEAN_UPS', 'COMPLIANCE', 'INITIAL']].sum().reset_index()
bbl_year_month.head()
```

```python
bbl_year_month['zip_year_month'] = bbl_year_month['zipcode'].astype(str)+' '+bbl_year_month['year_month'].astype(str)
bbl_year_month.head()
```

```python
bbl_year =df.groupby(by = ['zipcode', 'year'])[['BAIT', 'CLEAN_UPS', 'COMPLIANCE', 'INITIAL']].sum().reset_index()
bbl_year.head()
```

```python
zip_year_month = pd.merge(bbl_year_month, 
                          sightings, 
                          left_on=bbl_year_month['zip_year_month'], 
                          right_on=sightings['zip_year_month'], 
                          how='left')
#zip_year_month = zip_year_month.fillna(0)
#df['bbl_y'] = df['bbl_y'].astype(int)
#df['zipcode'] = df['zipcode'].astype(int)
zip_year_month.head()
```

```python
zip_year_month = zip_year_month[['zipcode', 'year_month_x', 'BAIT', 'CLEAN_UPS', 'COMPLIANCE', 'INITIAL', 'sighting']]
zip_year_month = zip_year_month.fillna(0)
zip_year_month.head()
```
